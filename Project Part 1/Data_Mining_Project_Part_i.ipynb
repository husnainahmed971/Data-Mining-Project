## Part 1:Utilize appropriate visualizations to show the distribution of features within the dataset. This includes histograms, density plots, boxplots, or other relevant graphs.

```python
numerical_features = ['user_followers', 'user_friends', 'user_favourites', 'retweets', 'favorites']
```

### Histogram: 

```python
# Display histograms for numerical features
df[numerical_features].hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

### Density plots

```python
# Create a figure and axis for subplots
fig, axes = plt.subplots(3, 2, figsize=(8, 8))
# Flatten the axis for easier iteration
axes = axes.flatten()
# Loop through numerical features and plot on subplots
for i, feature in enumerate(numerical_features):
  
    sns.kdeplot(df[feature], ax=axes[i])
    
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Density')
    axes[i].set_title(f'Density Plot of {feature}')
    
    

# Hide any unused subplots
for j in range(len(numerical_features), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

### Box Plot

```python
# Boxplots for numerical features
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_features])
plt.title('Boxplot of Numerical Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show
```

## Part 2: Investigate and visualize relationships between pairs of features. Use scatter plots or correlation matrices to illustrate correlations.

```python
# Scatter plot of user_followers and user_friends
plt.figure(figsize=(8, 6))
sns.scatterplot(x='user_followers', y='user_friends', data=df)
plt.title('Relationship between User Followers and User Friends')
plt.xlabel('User Followers')
plt.ylabel('User Friends')
plt.show()


```

```python
# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Create a correlation matrix
corr = df.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

# 2. Data Wrangling (20 marks):
## Perform essential data wrangling tasks, including handling missing values, duplicates, and outliers. Use appropriate techniques in Python (pandas) to ensure data cleanliness.

### Handling Missing Values

```python
df.isnull().sum()
```

```python
# Assuming 'df' is your DataFrame
sns.heatmap(df.isnull(), yticklabels=False, cmap='summer')
plt.show()
```

```python
# Check missing values
missing_values = df.isnull().sum()

# Fill missing numerical values with mean/median
df['user_verified'].fillna(df['user_verified'].median(), inplace=True)
df['user_location'].fillna('Unknown', inplace=True)
# Fill missing categorical/text values with the most frequent value
df['hashtags'].fillna(df['hashtags'].mode()[0], inplace=True)
df['user_description'].fillna('No description available', inplace=True)

```

### Output After Handling Missing Values

```python
# Assuming 'df' is your DataFrame
sns.heatmap(df.isnull(), yticklabels=False, cmap='summer')
plt.show()
```

### Handling Duplicates:


```python
# Check for duplicates
duplicates = df[df.duplicated()]
# Remove duplicates
df.drop_duplicates(inplace=True)
```

### Handling Outliers:


```python
# Visualize outliers using boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['user_followers', 'user_friends']])
plt.title('Boxplot of User Followers and User Friends')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

### Display after Handling OutLiers:

```python
# Outlier handling (example using IQR)
Q1 = df['user_followers'].quantile(0.25)
Q3 = df['user_followers'].quantile(0.75)
IQR = Q3 - Q1
# Remove outliers
df = df[~((df['user_followers'] < (Q1 - 1.5 * IQR)) | (df['user_followers'] > (Q3 + 1.5 * IQR)))]
```

```python
# Visualize outliers using boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['user_followers', 'user_friends']])
plt.title('Boxplot of User Followers and User Friends')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

# 3. Feature Engineering (25 marks):
## Engineer new features or transform existing ones to enhance the predictive power of the dataset, where it is possible/required. Consider techniques such as one-hot encoding, scaling, or creating interaction terms.

```python

# Assuming 'user_followers' and 'user_friends' are numerical columns
scaler = MinMaxScaler()
df[['user_followers', 'user_friends']] = scaler.fit_transform(df[['user_followers', 'user_friends']])
```

```python
# Creating an interaction term between 'user_followers' and 'user_friends'
df['followers_friends_interaction'] = df['user_followers'] * df['user_friends']
```

```python
# Creating an interaction term between 'user_followers' and 'user_friends'
df['followers_friends_interaction'] = df['user_followers'] * df['user_friends']
```

```python
# Extracting day, month, and year from the 'date' column
df['day'] = pd.to_datetime(df['date']).dt.day
df['month'] = pd.to_datetime(df['date']).dt.month
df['year'] = pd.to_datetime(df['date']).dt.year
```

For text-based features, techniques like tokenization or creating word embeddings can be valuable.

```python
#'text' is a text column
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
text_matrix = vectorizer.fit_transform(df['text'])
text_features = pd.DataFrame(text_matrix.toarray(), columns=vectorizer.get_feature_names_out())
df = pd.concat([df, text_features], axis=1)
```
